# Quonto

* To make sure the company stays profitable, they need to have an average default rate below 8.5%.

```python
# Quonto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
df = pd.read_csv("./assets/Python/qonto_loans.csv")
df.head()
```

```python
df.describe(include="all")
```

```python
# Define H1 et H0
H1 = "On a un default_rate < 8.5%"
H0 = "On a un default_rate >= 8.5%"

# Define a confidence level α
alpha = 0.05

# Calculate Z-Score
# Z = (X-µ)/s/sqrt(n)
mu = 8.5 / 100.0

Z = (df.TARGET.mean() - mu) / (df.TARGET.std() / len(df) ** 0.5)
print(Z)
```

```python
# Deduct p-value
from statsmodels.stats.weightstats import ztest

z_score, p_value = ztest(df.TARGET, alternative="smaller", value=mu)
print(p_value) 

X = norm(mu, (df.TARGET.std() / len(df) ** 0.5))
# Cumulative Distribution Function (CDF) 
p_value2 = X.cdf(df.TARGET.mean())  
print(p_value2) 


```

```python
# Compare p-value to confidence level
# Conclude (ou pas)
if (p_value<alpha):
  print(f"H0 rejetée donc H1 est acceptée")
  print(H1)
else:
  print(f"H0 ne peut pas être rejetée") 
  print(H0) 
```
