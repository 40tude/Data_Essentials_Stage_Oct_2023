```python
# Les données de l'exo
mean    =  .62
std     =  .49
z_score = 2.21

# Calcule la surface sous la courbe à droite du z_score
p = NormalDist(mean, std).cdf(z_score)
p = 1-p
print("p-value         :", p)

# Retrouve la valeur de x pour lequel la surf sous la courbe vaut 5%
Critical = NormalDist(mean, std).inv_cdf(0.95)
print("Valeur critique :", Critical)

# Crée une figure et un jeu d'axes
fig, ax = plt.subplots()

# Trace la courbe en bleu
# Génère des valeurs entre -1 et 2.5 espacées de .01
# Pour l'ensemble des x_values, génère les valeurs d'une gaussienne dont la moyenne est mean et l'écart type vaut std 
# pdf = Probability density function 
x_values = np.arange(-1, 2.5, 0.01)
y_values = norm.pdf(x_values, mean, std)        
ax.plot(x_values, y_values)
ax.fill_between(x_values, y_values, 0, color="blue", alpha=0.1)          

# Trace la courbe en vert
x_values3 = np.arange(Critical, 2.5, 0.01)
y_values3 = norm.pdf(x_values3, mean, std)
ax.plot(x_values3, y_values3, color="green", alpha=1.00)
ax.fill_between(x_values3, y_values3, 0, color="green", alpha=0.1)

# Trace la courbe en rouge
x_values2 = np.arange(z_score, 2.5, 0.01)
y_values2 = norm.pdf(x_values2, mean, std)
ax.plot(x_values2, y_values2, color="red", alpha=1.00)
_ = ax.fill_between(x_values2, y_values2, 0, color="red", alpha=0.1)


```
