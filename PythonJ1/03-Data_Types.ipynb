# Types of data

What are the data types of the following statistics?

1. Age
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Quantitative Discrete

2. Salary
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Quantitative continuous 

3. Country
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Qualitative Nominal

4. Color
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Qualitative Nominal

5. Sex
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Qualitative Nominal

6. Customer satisfaction with a service
    1. Quantitative / Qualitative?
    2. Ordinal / Nominal?
    3. Discrete / Continuous?

Réponse : Qualitative Ordinal
